```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
from datetime import datetime
from datetime import timedelta
import numpy as np
```

```python
str_filename = 'alanKi_output.csv'
```

```python
fh = open(str_filename)
csv_reader = csv.reader(fh)
```

```python
csv_header = next(csv_reader)
csv_header
```

```python
fh.close()
```

```python
df_sig = pd.read_csv(str_filename, header=None, skiprows=1, names=csv_header)
df_sig
```

```python
df_sig['Date'] = df_sig['DateTime'].map(lambda x: datetime.strptime(str(x), '%Y-%m-%d %H:%M:%S'))
df_sig['Date']
```

```python
mintStart = df_sig['Date'].iloc[0]
x = df_sig['Date'] - mintStart
x = x.dt.total_seconds()
y = df_sig['ID']
plt.plot(x, y)
plt.xlabel('Time(seconds)')
plt.ylabel('n minted')
```
